```python
# Video Generating function
#this works well
def generate_video():
    image_folder = '/home/theekshana/Documents/visualDet3D/workdirs/seq image 2011_10/warning/' # make sure to use your folder
    video_name = 'collision_warning.avi'
#     os.chdir("C:\\Python\\Geekfolder2")
      

    frame = cv2.imread(image_folder+"collision_warning 000.png")
  
    # setting the frame width, height width
    # the width, height of first image
    height, width, layers = frame.shape  
  
    video = cv2.VideoWriter(video_name, 0, 10, (width, height)) 
  
    # Appending the images to the video one by one
    for i in range(last_image_number+1):
        print(i)
        video.write(cv2.imread(image_folder+"collision_warning %03d.png"%i))
      
    # Deallocating memories taken for window creation
    cv2.destroyAllWindows() 
    video.release()  # releasing the video generated
  
    print('done!')
```

```python
# resize and rename collected local images
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output

image_folder='./local/local_0/right/'
write_path='./local/local_0_selected/right/'
start=2000
count=7481

#fig = plt.figure(figsize=(16,9))
    
for i in range(count):
    j=start+i
    image = cv2.imread(image_folder+"%06d"%j+".jpg")
        
    stretch_near = cv2.resize(image, (1242, 375),interpolation = cv2.INTER_LINEAR)
    cv2.imwrite( write_path+"%06d" % i+".png",stretch_near)
    
    img2 = stretch_near[:,:,::-1]
    clear_output(wait=True)
    plt.xlabel(i)
    plt.imshow(img2)
    plt.show()
```
